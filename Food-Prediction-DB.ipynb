```python
import pandas as pd 
df = pd.read_csv('continents_249.csv')
df2 = pd.read_csv('price_of_healthy_diet_clean.csv')

# Merge the two DataFrames on the 'country' column
merged_df = pd.merge(df, df2, on='country')

print(merged_df.head())
print(merged_df.columns)
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('food_prices.db')
cursor = conn.cursor()

cursor.execute("""
    CREATE TABLE IF NOT EXISTS food_data (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        continent TEXT,
        country TEXT,
        country_code TEXT,
        cost_healthy_diet_ppp_usd FLOAT
    )
""")

# Convert DataFrame to list of tuples
rows = list(merged_df[['continent', 'country', 'country_code', 'cost_healthy_diet_ppp_usd']].itertuples(index=False, name=None))

# Insert all at once
cursor.executemany("""
    INSERT INTO food_data (continent, country, country_code, cost_healthy_diet_ppp_usd)
    VALUES (?, ?, ?, ?)
""", rows)

conn.commit()

print(f"Inserted {len(rows)} rows in one batch!")

cursor.execute("SELECT * FROM food_data LIMIT 5")
results = cursor.fetchall()
for row in results:
    print(row)  
conn.close()
```

```python
import sqlite3
import pandas as pd

df_combined = pd.read_csv("combined_yearly.csv")

conn = sqlite3.connect('data.db')
cursor = conn.cursor()

# Enable foreign key support in SQLite (off by default!)
cursor.execute("PRAGMA foreign_keys = ON")

# Create countries table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS countries (
        country_code INTEGER PRIMARY KEY,
        continent TEXT,
        country TEXT
    )
""")

# Create datatable with correct foreign key syntax
cursor.execute("""
    CREATE TABLE IF NOT EXISTS datatable (
        food_id INTEGER PRIMARY KEY AUTOINCREMENT,
        country_code INTEGER,
        year FLOAT,
        commodity TEXT,
        cost_healthy_diet_ppp_usd FLOAT,
        annual_cost_healthy_diet_usd FLOAT,
        cost_vegetables_ppp_usd FLOAT,
        cost_fruits_ppp_usd FLOAT,
        total_food_components_cost FLOAT,
        FOREIGN KEY (country_code) REFERENCES countries(country_code)
    )
""")

# Insert into countries first (parent table must be populated before child)
columns1 = ['country_code', 'continent', 'country']
country_rows = list(
    df_combined[columns1]
    .drop_duplicates(subset='country_code')  # avoid duplicate country entries
    .itertuples(index=False, name=None)
)
cursor.executemany("""
    INSERT OR IGNORE INTO countries (country_code, continent, country)
    VALUES (?, ?, ?)
""", country_rows)

# Insert into datatable (child table)
columns2 = [
    'country_code',
    'year',
    'commodity',
    'cost_healthy_diet_ppp_usd',
    'annual_cost_healthy_diet_usd',
    'cost_vegetables_ppp_usd',
    'cost_fruits_ppp_usd',
    'total_food_components_cost'
]
rows = list(df_combined[columns2].itertuples(index=False, name=None))
cursor.executemany("""
    INSERT INTO datatable (country_code, year, commodity, cost_healthy_diet_ppp_usd,
                           annual_cost_healthy_diet_usd, cost_vegetables_ppp_usd,
                           cost_fruits_ppp_usd, total_food_components_cost)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
""", rows)

conn.commit()
print(f"Inserted {len(country_rows)} countries and {len(rows)} data rows!")

cursor.execute("SELECT * FROM datatable LIMIT 5")
for row in cursor.fetchall():
    print(row)

conn.close()
```
