```python
import pandas as pd 
df = pd.read_csv('continents_249.csv')
df2 = pd.read_csv('price_of_healthy_diet_clean.csv')

# Merge the two DataFrames on the 'country' column
merged_df = pd.merge(df, df2, on='country')

print(merged_df.head())
print(merged_df.columns)
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('food_prices.db')
cursor = conn.cursor()

cursor.execute("""
    CREATE TABLE IF NOT EXISTS food_data (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        continent TEXT,
        country TEXT,
        country_code TEXT,
        cost_healthy_diet_ppp_usd FLOAT
    )
""")

# Convert DataFrame to list of tuples
rows = list(merged_df[['continent', 'country', 'country_code', 'cost_healthy_diet_ppp_usd']].itertuples(index=False, name=None))

# Insert all at once
cursor.executemany("""
    INSERT INTO food_data (continent, country, country_code, cost_healthy_diet_ppp_usd)
    VALUES (?, ?, ?, ?)
""", rows)

conn.commit()

print(f"Inserted {len(rows)} rows in one batch!")

cursor.execute("SELECT * FROM food_data LIMIT 5")
results = cursor.fetchall()
for row in results:
    print(row)  
conn.close()
```
